```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/pd_speech_features.csv', index_col = False)
df.head()
```

```python
df.info()
```

```python
dic = {}
for i in df.columns:
  dic[df[i][0]] = df[i][1:].astype(float)
  
data = pd.DataFrame(dic)
data.head()
```

```python
print(data.columns)
data['PPE'].dtype
data.info()
```

```python
data.drop('id', axis = 1, inplace=True)
data.head()
```

```python
from sklearn.model_selection import train_test_split
y = data['class']
X = data.drop('class', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=102, test_size = 0.3)
```

```python
print(y_train.value_counts(), y_train.count())
print(y_test.value_counts(), y_test.count())
```

## Quetion: what would be the test/train accuracy if we had a model that would only ouput one?

##### Well sicne we have 391 ones out of 529 records in our train set and 173 ones out of 227 records in out test set:
*    Train_Accuracy : 0.73913
*    Test_Accuracy : 0.76211

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# ✅Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
dtc_param_dic = {
    "max_depth" : [2,3,4,5,8,10,15,20],
    "min_samples_split" : [1,2,3,45,8,10]
}
grid = GridSearchCV(dtc, 
                    param_grid = dtc_param_dic,
                    cv = 3,
                    verbose = 1,
                    n_jobs = -1)
grid.fit(X_train, y_train)

print("best parameters: ", grid.best_params_)
print("best estimator: ", grid.best_estimator_)
print("best score: ", grid.best_score_)
```

```python
dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(X_train, y_train)
y_pred_train_dtc = dtc.predict(X_train)
y_pred_test_dtc = dtc.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_train_dtc = accuracy_score(y_train, y_pred_train_dtc)
accuracy_test_dtc = accuracy_score(y_test, y_pred_test_dtc)
conf = confusion_matrix(y_test, y_pred_test_dtc)
clf = classification_report(y_test, y_pred_test_dtc)
accuracy_test_dtc
```

```python
from sklearn.model_selection import cross_val_score
scores_dtc = cross_val_score(dtc, X_train, y_train, cv=5)
print(scores_dtc.mean())
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(dtc, X_test, y_pred_test_dtc)
```

## First of all, we should consider that in general, applying PCA before building a model will NOT help to make the model perform better (in terms of accuracy)! This is because PCA is an algorithm that does not consider the response variable / prediction target into account. PCA will treat the feature has large variance as important features, but the feature has large variance can have noting to do with the prediction target.
## So the goal is to get a result which is close to the one before PCA but this time with fewer features. A we compare this table to the one before PCA we can see that the drop in accuracy is not too much but this time we get these results with only 60 feature. Going from 750 features to 60 with almost same result is a success.
